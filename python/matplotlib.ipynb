### histogram

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
data = np.random.randn(1000)  # Normally distributed random data

# Create a histogram
plt.hist(data, label='bins=10')
plt.hist(data, bins=30, color='skyblue', label='bins=30')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')

# Show the plot
plt.show()
```

```python
# Generate some random data
data = np.random.randn(1000)  # Normally distributed random data

# Create a histogram
plt.hist(data, bins=30, color='skyblue', label='bins=30', cumulative=True)
plt.hist(data, label='bins=10')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')
plt.legend(loc='upper left')

# Show the plot
plt.show()
```

```python
# Generate some random data
data = np.random.randn(1000)  # Normally distributed random data
data2 = np.random.randn(1000)  # Normally distributed random data

# Create a histogram
plt.hist((data, data2), bins=30, label='bar', histtype='bar')
plt.figure()

plt.hist((data, data2), bins=30, label='barstacked', histtype='barstacked')
plt.figure()

plt.hist((data, data2), bins=30, label='step', histtype='step')
plt.figure()

plt.hist((data, data2), bins=30, label='stepfilled', histtype='stepfilled')
plt.figure()

# Show the plot
plt.show()
```

```python
# standard deviation = 2.0, mean = 1.0, normal distribution
a = 2.0 * np.random.randn(10000) + 1.0
# standard normal distribution
b = np.random.standard_normal(10000)
# random values between -10.0 ~ 10.0
c = 20.0 * np.random.rand(5000) - 10.0

plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step')
plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled')
plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step')

plt.show()
```
