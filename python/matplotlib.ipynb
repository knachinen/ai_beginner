### Line style

```python
plt.plot([1,2,3], [4,4,4], '-', color='C0', label='Solid')
plt.plot([1,2,3], [3,3,3], '--', color='C0', label='Dashed')
plt.plot([1,2,3], [2,2,2], ':', color='C0', label='Dotted')
plt.plot([1,2,3], [1,1,1], '-.', color='C0', label='Dash-dot')

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4)
plt.show()
```

```python
plt.plot([1,2,3], [4,4,4], linestyle=(0, (1, 1)), color='C0', label='(0, (1, 1))')
plt.plot([1,2,3], [3,3,3], linestyle=(0, (1, 5)), color='C0', label='(0, (1, 5))')
plt.plot([1,2,3], [2,2,2], linestyle=(0, (5, 1)), color='C0', label='(0, (5, 1))')
plt.plot([1,2,3], [1,1,1], linestyle=(0, (3, 5, 1, 5)), color='C0', label='(0, (3, 5, 1, 5))')

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4)
plt.show()
```

```python
plt.plot([1,2,3], [4,4,4], linestyle='solid', linewidth=10, 
         solid_capstyle='butt', color='C0', label='solid+butt')
plt.plot([1,2,3], [3,3,3], linestyle='solid', linewidth=10,
         solid_capstyle='round', color='C0', label='solid+round')
plt.plot([1,2,3], [2,2,2], linestyle='dashed', linewidth=10,
         solid_capstyle='butt', color='C1', label='dashed+butt')
plt.plot([1,2,3], [1,1,1], linestyle='dashed', linewidth=10,
         solid_capstyle='round', color='C0', label='dashed+round')

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4)
plt.show()
```

### fill

```python
x = [1, 2, 3, 4]
y = [2, 3, 5, 10]

plt.plot(x, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_between(x[1:3], y[1:3], alpha=0.5)

plt.show()
```

```python
plt.plot(x, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_betweenx(y[2:4], x[2:4], alpha=0.5)

plt.show()
```

```python
y2 = [1, 2, 4, 8]

plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_between(x[1:3], y[1:3], y2[1:3], color='lime', alpha=0.5)

plt.show()
```

```python
plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill([1.9, 1.9, 3.1, 3.1], [1., 4., 6., 3.], color='cyan', alpha=.5)

plt.show()
```

### axis scale

```python
import numpy as np

x = np.linspace(-10, 10, 100)
y = x ** 3

plt.plot(x, y)
# symmetrical log scale
plt.xscale('symlog')

plt.show()
```

```python
x = np.linspace(0, 5, 100)
y = np.exp(x)

plt.plot(x, y)
plt.yscale('linear')

plt.show()

plt.plot(x, y)
plt.yscale('log')

plt.show()
```

### horizontal/vertical line

```python
x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axhline(4.0, 0.1, 0.9, color='lightgray', linestyle='--', linewidth=2)
plt.hlines(-0.62, 1.0, 2.5, color='gray', linestyle='solid', linewidth=3)

plt.show()
```

```python
x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axvline(.1, 0.1, 0.9, color='lightgray', linestyle='--', linewidth=2)
plt.hlines(-0.62, 1.0, 2.5, color='gray', linestyle='solid', linewidth=3)

plt.show()
```
