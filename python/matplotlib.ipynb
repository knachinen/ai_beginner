```python
import matplotlib.pyplot as plt
```

```python
val = [1, 2, 3, 4]

plt.plot(val)
plt.show()
```

```python
val2 = [2**1, 2**2, 2**3, 2**4]

# x-y
plt.plot(val, val2)
plt.show()
```

- ro
    - r: red color
    - o: o-shape marker
- b-
    - b: blue color
    - -: line

```python
# style

plt.plot(val, val2, 'ro')  # red color, 'o' marker
plt.axis([0, 5, 0, 20])
plt.show()
```

```python
import numpy as np
```

```python
t = np.arange(0., 5., 0.2)

plt.plot(t, t, 'r--', 
         t, t**2, 'bs', 
         t, t**3, 'g^',
         t, t**4, 'yx'
        )
plt.show()
```

```python
data_dict = {'data_x': np.arange(1, 6, 1),
             'data_y': [2, 3, 5, 10, 8]
            }

plt.plot('data_x', 'data_y', data=data_dict)
plt.show()
```

```python
plt.plot(val, val2)

# setting x, y labels
plt.xlabel('X-Label')
plt.ylabel('Y-Label')

plt.show()
```

```python
plt.plot(val, val2)

# padding
plt.xlabel('X-Label', labelpad=15)
plt.ylabel('Y-Label', labelpad=20)

plt.show()
```

```python
# setting font
font1 = {'family': 'serif',
         'color': 'b',
         'weight': 'bold',
         'size': 14
         }

font2 = {'family': 'fantasy',
         'color': 'deeppink',
         'weight': 'normal',
         'size': 'xx-large'
         }


plt.xlabel('X-Label', labelpad=15, fontdict=font1)
plt.ylabel('Y-Label', labelpad=20, fontdict=font2)

plt.show()
```

```python
plt.plot(val, val2)

# setting location
plt.xlabel('X-Label', labelpad=15, fontdict=font1, loc='left')
plt.ylabel('Y-Label', labelpad=20, fontdict=font2, loc='bottom')

plt.show()
```

```python
plt.plot(val, val2, label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

# legend
plt.legend()

plt.show()
```

```python
plt.plot(val, val2, label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

# legend location
plt.legend(loc=(.5, .5))

plt.show()
```

```python
plt.plot(val, val2, label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

# legend location
plt.legend(loc='lower right')

plt.show()
```

```python
plt.plot(val, val2, label='graph 1')
plt.plot(val2, val1, label='graph 2')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

# legend
plt.legend(loc='best', ncol=2)

plt.show()
```
